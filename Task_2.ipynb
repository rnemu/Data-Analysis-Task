```python
# import library to connect the database
import mysql.connector as c
mydb = c.connect(host="localhost",
  user="root",
  password="Rakesh@123",database='practice'
)
mycursor = mydb.cursor()
# To execute sql query use execute
mycursor.execute('DROP TABLE IF EXISTS demo_bike;')
mycursor.execute("CREATE TABLE demo_bike (bike_name varchar(50),price  int,city  varchar(25),kms_driven  float,owner varchar(25), age  int,power float, brand varchar(25))")
```

```python
# insert the csv file into mysql
import csv
file = open('used_bikes.csv')
csv_data = csv.reader(file)
skipHeader = True

for row in csv_data:
    if skipHeader:
        skipHeader = False
        continue
    
    mycursor.execute("INSERT INTO demo_bike(bike_name, price, city, kms_driven, owner, age, power,brand) values(%s,%s,%s,%s,%s,%s,%s,%s)",row)
mydb.commit()    
```
